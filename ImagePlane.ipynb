```python
import matplotlib.pyplot as plt
import numpy as np
from skimage import io
```

```python
plane = io.imread('plane.png', format='png')
plt.imshow(plane)
plt.show()
plane.shape
```

```python
#Une mthode qui rtourne la figure et l'axe
#subplots -> mthode de matplolib qui rtourne l'image et l'axe
#Puis on donne un nom  cet image
def figAxe(img, Imgname):
    fig, ax = plt.subplots()
    ax.set_title(Imgname)
    ax.axis('off') #Masque les axes
    ax.imshow(img)
    return fig, ax
```

```python
fig, ax = figAxe(plane, 'A plane\n')
```

```python
rows = plane.shape[0]
cols = plane.shape[1]
print(f"Lignes : {rows}, Colonnes : {cols}")
```

```python
zoom = plane[50:-100, 40:-140]

fig1, ax1 = figAxe(zoom, 'Image de plane zoomer\n')
```

```python
zoom[zoom < 50] = 100
fig1, ax1 = figAxe(zoom, 'On ajuste un peu la luminosit\n')
```

```python
zoom = zoom[::2, ::2]
fig1, ax1 = figAxe(zoom, 'Un peu de filtre\n')
```
